```python
apakah keadaan cuaca mempengaruhi jumlah rental sepeda? dan apakah ada perbedaan antara hari kerja atau tidak?
```

```python
sns.boxplot(y='cnt', x='weathersit', hue='workingday', data=hour_df)
title = 'PENGARUH CUACA DAN HARI KERJA TERHADAP RENTAL SEPEDA'
plt.title(title)
plt.xlabel("Keadaan Cuaca")
plt.ylabel("Jumlah Rental")
xlabels = ['Cerah', 'Mendung', 'Hujan Ringan', 'Badai']
x = np.arange(len(xlabels))
plt.xticks(x, labels=xlabels)
L = plt.legend()
L.get_texts()[0].set_text('Hari Libur')
L.get_texts()[1].set_text('Hari Kerja')
plt.show()
```

```python
sns.boxplot(y='cnt', x='season',hue='workingday',data=hour_df)
title="PENGARUH MUSIM TERHADAP JUMLAH RENTAL SEPEDA"
plt.title=title
plt.xlabel("Musim")
plt.ylabel("Jumlah Rental")
xlabels=['Semi','Panas','Gugur','Dingin']
x=np.arange(len(xlabels))
plt.xticks(x, labels=xlabels)
L = plt.legend()
L.get_texts()[0].set_text('Hari Libur')
L.get_texts()[1].set_text('Hari Kerja')
plt.show()
```
